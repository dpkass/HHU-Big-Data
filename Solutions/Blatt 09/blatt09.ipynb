```python
import os
import pandas as pd

from tqdm.notebook import tqdm
```

```python
def read_dir(dir):
    files = os.listdir(dir)
    files = [os.path.join(dir, file) for file in files]
    dfs = [pd.read_parquet(file) for file in tqdm(files, desc='Reading files')]
    return pd.concat(dfs)


base_dir = './data/nyc_tlc_trip_record_2023'

get_yellow = lambda: read_dir(base_dir + '/yellow')
get_green = lambda: read_dir(base_dir + '/green')
get_fhv = lambda: read_dir(base_dir + '/fhv')
get_fhvhv = lambda: read_dir(base_dir + '/fhvhv')
```
